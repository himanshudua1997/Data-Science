```python
#importing postal data of Toronto
postal_df=pd.read_csv('post.csv')
```

```python
#merging Geospatial_Coordinates and postal dataframe into single dataframe
result = pd.merge(left=postal_df,right=canada_coor,left_on='Postcode',right_on='Postal Code')
```

```python
#removing extra postal code column
result.drop('Postal Code',axis=1)
```

```python
import geopy
import folium
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe
```

```python
address = 'Toronto'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of Toronto using latitude and longitude values
map_newyork = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, Borough, Neighborhood in zip(result['Latitude'], result['Longitude'],result['Borough'], result['Neighbourhood']):
    label = '{}, {}'.format(Neighborhood, Borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_newyork)  
    
map_newyork
```
