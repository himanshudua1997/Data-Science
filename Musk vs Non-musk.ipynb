```python
import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("musk_csv.csv")
```

```python
df.head()
```

```python
X = df.iloc[:,3:171]
```

```python
X
```

```python
Y = df[['class']]
```

```python
X = np.array(X)
```

```python
Y = np.array(Y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state = 1)
```

```python
from sklearn import preprocessing
```

```python
X_train= preprocessing.StandardScaler().fit(X_train).transform(X_train)
X_test= preprocessing.StandardScaler().fit(X_test).transform(X_test)
X_train.shape
```

```python
X_test.shape
```

```python
Y_train.shape
```

```python
Y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

```python
LR = LogisticRegression(C=0.01, solver='liblinear').fit(X_train,Y_train)
LR
```

```python
pred = LR.predict(X_test)
```

```python
from sklearn import metrics
from sklearn.metrics import log_loss
```

```python
y_hat=LR.predict_proba(X_test)
pred_log_loss = metrics.log_loss(Y_test,y_hat)
```

```python
pred_log_loss
```

```python
from sklearn.metrics import jaccard_similarity_score
from sklearn.metrics import f1_score
from sklearn.metrics import log_loss
```

```python
LR_MODEL=LR.predict(X_test)
LR_MODEL_p= LR.predict_proba(X_test)
print("jaccard_similarity_score")
print(jaccard_similarity_score(Y_test, LR_MODEL))
print("f1_score")
print(f1_score(Y_test, LR_MODEL,average='weighted'))
print("LogLoss")
print(log_loss(Y_test, LR_MODEL_p))
```
